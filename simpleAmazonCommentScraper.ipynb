```python
link = 'https://www.amazon.com.tr/Columbia-Peakfreak-Outdry-y%C3%BCr%C3%BCy%C3%BC%C5%9F-ayakkab%C4%B1s%C4%B1/product-reviews/B08YS8X61G/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
urls = rq.get(link).text
```

```python
urls
```

```python
soup = bs(urls,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
comment = soup.find_all("span",{"data-hook":"review-body"})
```

```python
names 
```

```python
cust_name = []
for i in range(2,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
comment
```

```python
comment_content = []
for i in range(0,len(comment)):
    comment_content.append(comment[i].get_text())
comment_content
```

```python
comment_content[:] = [comments.lstrip('\n') for comments in comment_content]
comment_content
```

```python
comment_content[:] = [comments.rstrip('\n') for comments in comment_content]
comment_content
```

```python
comment_content[:] = [comments.replace("  ", "") for comments in comment_content]
comment_content
```

```python
df = pd.DataFrame()
```

```python
df['names']=cust_name
```

```python
df['comments']=comment_content
```
